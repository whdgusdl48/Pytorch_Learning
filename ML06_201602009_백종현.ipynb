```
from sklearn.datasets import make_moons

X,y = make_moons(n_samples = 200, noise = 0.05, random_state = 0)
```

```
import matplotlib.pyplot as plt

plt.scatter(X[:,0],X[:,1])
plt.savefig('basic.png')  
```

```
plt.scatter(X[:,0],X[:,1],c=y)
plt.title('answer')
plt.savefig('Answer.png') 
```

```
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters = 2)
kmeans.fit(X)
y_pred = kmeans.predict(X)
plt.title('Kmeans Clustering 2')
plt.scatter(X[:,0],X[:,1],c=y_pred, s=60, edgecolors='k')
plt.savefig('KmeansCluster2.png')
```

```
plt.figure(figsize=(20,12))
for i in range(2,11):
  plt.subplot(3,3,i-1)
  kmeans = KMeans(n_clusters = i)
  kmeans.fit(X)
  y_pred = kmeans.predict(X)
  plt.title('kmeans cluster '+str(i))
  plt.scatter(X[:,0],X[:,1],c=y_pred, s=60, edgecolors='k')
plt.savefig('Kmeans.png')  
```

```
from sklearn.datasets import make_blobs

# 0.5, 1 ,2.5
plt.figure(figsize=(16,10))
plt.subplot(1,2,1)
plt.title("Answer")
X, y= make_blobs(n_samples=200,cluster_std=[0.5,1,2.5], centers=3, random_state=1)
plt.scatter(X[:, 0], X[:, 1], marker='o', s=100,
edgecolor="k", linewidth=2)
plt.xlabel("$X_1$")
plt.ylabel("$X_2$")
plt.subplot(1,2,2)
plt.title("Kmeans_predict")
y_pred = KMeans(n_clusters=3, random_state=0).fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], marker='o', c=y_pred, s=100,
edgecolor="k", linewidth=2)
plt.xlabel("$X_1$")
plt.ylabel("$X_2$")
plt.show()
plt.savefig('fig2.png')  
```
