```
from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(data2, test_size= 0.2, random_state = 42)
```

```
train_set
```

```
train_y = train_set['race/ethnicity']
del train_set['race/ethnicity']
```

```
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=4)
knn.fit(test_set, test_y)
```

```
some_data = test_set.iloc[6:10]
some_labels = test_y.iloc[6:10]
print("예측:", knn.predict(some_data))
print("레이블:", list(some_labels))
```

```
from sklearn.metrics import mean_squared_error

predictions = knn.predict(train_set)
lin_mse = mean_squared_error(train_y, predictions)
lin_rmse = np.sqrt(lin_mse)
lin_rmse
```

```
from sklearn.tree import DecisionTreeClassifier

tree_reg = DecisionTreeClassifier(random_state=42)
tree_reg.fit(train_set, train_y)
```

```
some_data = train_set.iloc[6:10]
some_labels = train_y.iloc[6:10]
print("예측:", tree_reg.predict(some_data))
print("레이블:", list(some_labels))
```

```
predictions = tree_reg.predict(train_set)
lin_mse = mean_squared_error(train_y, predictions)
lin_rmse = np.sqrt(lin_mse)
lin_rmse
```

```
from sklearn.model_selection import cross_val_score

scores = cross_val_score(tree_reg, train_set, train_y,
                         scoring="neg_mean_squared_error", cv=10)
tree_rmse_scores = np.sqrt(-scores)
def display_scores(scores):
    print("점수:", scores)
    print("평균:", scores.mean())
    print("표준 편차:", scores.std())

display_scores(tree_rmse_scores)
```

```
from sklearn.model_selection import cross_val_score

scores = cross_val_score(knn, train_set, train_y,
                         scoring="neg_mean_squared_error", cv=10)
tree_rmse_scores = np.sqrt(-scores)
def display_scores(scores):
    print("점수:", scores)
    print("평균:", scores.mean())
    print("표준 편차:", scores.std())

display_scores(tree_rmse_scores)
```

```
from sklearn.ensemble import RandomForestClassifier

forest_reg = RandomForestClassifier(n_estimators=100, random_state=42)
forest_reg.fit(train_set, train_y)
```

```
some_data = train_set.iloc[6:10]
some_labels = train_y.iloc[6:10]
print("예측:", forest_reg.predict(some_data))
print("레이블:", list(some_labels))
```

```
scores = cross_val_score(forest_reg, train_set, train_y,
                         scoring="neg_mean_squared_error", cv=10)
tree_rmse_scores = np.sqrt(-scores)
def display_scores(scores):
    print("점수:", scores)
    print("평균:", scores.mean())
    print("표준 편차:", scores.std())

display_scores(tree_rmse_scores)
```

```
from sklearn.model_selection import GridSearchCV
# 모델을 세부튜닝을 시작합니다.
# 만족할만한 하이퍼파라미터의 조합을 찾아서 수동으로 돌리도록합니다. 탐욕알고리즘을 통해서 찾습니다.
param_grid = [
    # 12(=3×4)개의 하이퍼파라미터 조합을 시도합니다.
    {'n_estimators': [30, 100, 300], 'max_features': [2, 3, 4, 6]},
    # bootstrap은 False로 하고 6(=2×3)개의 조합을 시도합니다.
    {'bootstrap': [False], 'n_estimators': [3, 10], 'max_features': [2, 3, 4]},
  ]
# 다섯 개의 폴드로 훈련하면 총 (12+6)*5=90번의 훈련이 일어납니다.
grid_search = GridSearchCV(forest_reg, param_grid, cv=5,
                           scoring='neg_mean_squared_error',
                           return_train_score=True)
grid_search.fit(train_set, train_y)
```

```

cvres = grid_search.cv_results_
for mean_score, params in zip(cvres["mean_test_score"], cvres["params"]):
    print(np.sqrt(-mean_score), params)
```

```
y_pred = forest_reg.predict(train_set)
print("테스트 세트의 정확도: {:.2f}".format(np.mean(y_pred == train_y)))
```
