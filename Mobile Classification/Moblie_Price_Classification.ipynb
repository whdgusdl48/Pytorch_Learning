```
train_x = data_copy[attributes]
train_y = train_x['price_range']
del train_x['price_range']
```

```
test = pd.read_csv('test.csv')
test_copy = test.copy()
```

```
test_copy
```

```
test_x = test_copy[['ram','battery_power','px_width','px_height']]
```

```
from sklearn.tree import DecisionTreeClassifier

tree_clf = DecisionTreeClassifier(max_depth = 4)

tree_clf.fit(train_x,train_y)
```

```
some_data = train_x.iloc[6:10]
some_labels = train_y.iloc[6:10]
print("예측:", tree_clf.predict(some_data))
print("레이블:", list(some_labels))
```

```
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=124)
knn.fit(train_x, train_y)
```

```
some_data = train_x.iloc[6:10]
some_labels = train_y.iloc[6:10]
print("예측:", knn.predict(some_data))
print("레이블:", list(some_labels))
```

```
from sklearn.ensemble import RandomForestClassifier

forest_clf = RandomForestClassifier(n_estimators=5, random_state=42)
forest_clf.fit(train_x, train_y)
```

```
some_data = train_x.iloc[6:10]
some_labels = train_y.iloc[6:10]
print("예측:", forest_clf.predict(some_data))
print("레이블:", list(some_labels))
```

```
from sklearn.metrics import mean_squared_error
predictions = tree_clf.predict(train_x)
tree_mse = mean_squared_error(train_y, predictions)
tree_rmse = np.sqrt(tree_mse)
tree_rmse
```

```
predictions = knn.predict(train_x)
knn_mse = mean_squared_error(train_y, predictions)
knn_rmse = np.sqrt(knn_mse)
knn_rmse
```

```
print("결정트리 훈련 세트의 정확도:",tree_clf.score(train_x,train_y))
print("랜덤포레스트 훈련 세트의 정확도:",forest_clf.score(train_x,train_y))
print('Kneighbor 훈련 세트의 정확도 : ', knn.score(train_x,train_y))

test_y = tree_clf.predict(test_x)
print("결정트리 테스트 세트의 정확도:",tree_clf.score(test_x,test_y))
test_y = forest_clf.predict(test_x)
print("랜덤포레스트 테스트 세트의 정확도:",forest_clf.score(test_x,test_y))
test_y = knn.predict(test_x)
print('Kneighbor 테스트 세트의 정확도 : ', knn.score(test_x,test_y))
```
